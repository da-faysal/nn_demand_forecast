```python
# Import required modules
import pandas as pd
import numpy as np
from itertools import repeat
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
# Set the random seed to make the result reproducible
torch.manual_seed(43)
```

```python
# Read in the data, fill the gaps. 7DA is abailabe for current day and 6 days ahead. So we can use it to make future forecast 
df = pd.read_csv("caiso_system_demand.csv", parse_dates=["time"]).set_index("time")
df.system_demand = df.system_demand.interpolate()
df.demand_7da = df.demand_7da.interpolate()
df.temp = df.temp.interpolate()

# Create hour and day
df["hour"] = df.index.hour
df["day"] = df.index.day
df.head()
```

```python
# Use min-max scaling
def min_max_scale(df, column):
    temp_df = df.copy(deep=True)
    temp_df[column] = (df[column]-df[column].min())/(df[column].max()-df[column].min())
    return temp_df[column]
```

```python
# Apply scaling
df_scaled = pd.concat(list(map(min_max_scale, repeat(df), ["system_demand", "demand_7da", "temp", "hour"])), axis=1)
df_scaled.head()
```

```python
# We will use last 7 days as test data, and remaining as train data
train = df_scaled.iloc[:-7*24].to_numpy().astype(np.float32)
test = df_scaled.iloc[-7*24:].to_numpy().astype(np.float32)

x_train = train[:, 1:]
y_train = train[:, 0].reshape(-1, 1)

x_test = test[:, 1:]
y_test = test[:, 0].reshape(-1, 1)

# Create tensor dataset
train_dataset = TensorDataset(torch.from_numpy(x_train), torch.from_numpy(y_train))
test_dataset = TensorDataset(torch.from_numpy(x_test), torch.from_numpy(y_test))

# Create data loader. Set batch size to 24
train_loader = DataLoader(train_dataset, batch_size=24, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=24, shuffle=False)
```

```python
# Define the multi-layer perceptron
class MLP(nn.Module):
    def __init__(self, in_features, out_features, hidd_act):
        super().__init__()
        
        self.layers = nn.Sequential(
            nn.Linear(in_features=in_features, out_features=out_features),
            hidd_act,
            nn.Linear(in_features=out_features, out_features=out_features//2),
            hidd_act,
            nn.Linear(in_features=out_features//2, out_features=1)
        )
        
    def forward(self, x):
        return self.layers(x)
```

```python
# Define the model parameters 
in_features = x_train.shape[-1] # Number of input features
hidden_size = 512 # Size of the hidden layer
hidd_act = nn.ReLU() # Activation function of hidden layer
lr = 1e-4 # learning rate

# Initialize the network
model = MLP(in_features=in_features, out_features=hidden_size, hidd_act=hidd_act)

# Define optimizer and loss func
optimizer = torch.optim.Adam(model.parameters(), lr=lr) 
loss_fn = nn.MSELoss()
```

```python
EPOCHS = 70

# Store epoch loss (avg loss)
train_loss = []
test_loss = []

for ep in range(1, EPOCHS+1):
    model.train()
    run_train_loss = 0
    run_test_loss = 0
    
    for inputs, target in train_loader:
        optimizer.zero_grad()
        output = model(inputs)
        loss = loss_fn(output, target)
        loss.backward()
        optimizer.step()
        
        # Calculate running loss
        run_train_loss += loss.item()*inputs.size(0)
        
    
    # Calculate avg loss (epoch loss)
    avg_loss_train = run_train_loss/len(train_loader.dataset)
    
    # Store the loss
    train_loss.append(avg_loss_train)
    
    # Inference on test data
    with torch.no_grad():
        for inputs, target in test_loader:
            output = model(inputs)
            loss = loss_fn(output, target)
            
            # Running loss
            run_test_loss += loss.item()*inputs.size(0)
        
        # Calculate avg loss (epoch loss)
        avg_loss_test = run_test_loss/len(test_loader.dataset)
        
        # Store test losse
        test_loss.append(avg_loss_test)
    
    # Print stats after every 10 epochs
    if ep%10==0:
        print(f"Epoch:{ep}, train loss: {train_loss[-1]}, test loss: {test_loss[-1]}")

# Create a df off losses and epoch
loss_df = pd.DataFrame({
    "train_loss":train_loss,
    "test_loss":test_loss,
    "epoch":range(1, EPOCHS+1)
})
```
